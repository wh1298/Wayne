# Import Packages

```python
# Common imports
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import numpy as np
import pandas as pd
import os
%matplotlib inline
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

# Import Data

```python
import pandas as pd

# Import dataset
fileName = "MultipleLinearRegression.csv"
a3 = pd.read_csv(fileName)
a3
```

# Plot the Data Points

```python
from mpl_toolkits import mplot3d

# Extract the data from the DataFrame
zdata = a3['z']
xdata = a3['x']
ydata = a3['y']

# Create a 3D figure with 2x2 subplots
fig = plt.figure(figsize=[20,20])

# Subplot 1
# Set axis labels and font color
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
ax1.set_xlabel('y', color='maroon')
ax1.set_ylabel('x', color='maroon')
ax1.set_zlabel('z', color='maroon')
# Change viewing angle
ax1.view_init(0, 180)
# Plot the data
ax1.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)

# Subplot 2
# Set axis labels and font color
ax2 = fig.add_subplot(2, 2, 2, projection='3d')
ax2.set_xlabel('x', color='maroon')
ax2.set_ylabel('y', color='maroon')
ax2.set_zlabel('z', color='maroon')
# Change viewing angle
ax2.view_init(37, 0)
# Plot the data
ax2.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)

# Subplot 3
# Set axis labels and font color
ax3 = fig.add_subplot(2, 2, 3, projection='3d')
ax3.set_xlabel('x', color='maroon')
ax3.set_ylabel('y', color='maroon')
ax3.set_zlabel('z', color='maroon')
# Change viewing angle
ax3.view_init(37, 45)
# Plot the data
ax3.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)

# Subplot 4
# Set axis labels and font color
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
ax4.set_xlabel('x', color='maroon')
ax4.set_ylabel('y', color='maroon')
ax4.set_zlabel('z', color='maroon')
# Change viewing angle
ax4.view_init(20, 26)
# Plot the data
ax4.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)
```

# Train Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

# Extract the features (x and y) and the response variable (z) from the DataFrame
X = a3[['x', 'y']]
y = a3['z']

# Instantiate a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

# Plot the Line of Best Fit

```python
# Plot Curve Fit
x_fit = np.linspace(0,21,1000)
y_fit = x_fit

# Create a DataFrame with x_fit and y_fit as features
fit_data = pd.DataFrame({'x': x_fit, 'y': y_fit})

# Use the model to predict z_fit values
z_fit = model.predict(fit_data)# Create a 3D figure with 2x2 subplots
fig = plt.figure(figsize=[20,20])

# Subplot 1
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
# Set axis labels and font color
ax1.set_xlabel('y', color='maroon')
ax1.set_ylabel('x', color='maroon')
ax1.set_zlabel('z', color='maroon')
# Change viewing angle
ax1.view_init(0, 180)
# Plot the data with the fit line
ax1.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)
ax1.plot3D(x_fit, y_fit, z_fit, 'black')

# Subplot 2
ax2 = fig.add_subplot(2, 2, 2, projection='3d')
# Set axis labels and font color
ax2.set_xlabel('x', color='maroon')
ax2.set_ylabel('y', color='maroon')
ax2.set_zlabel('z', color='maroon')
# Change viewing angle
ax2.view_init(37, 0)
# Plot the data with the fit line
ax2.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)
ax2.plot3D(x_fit, y_fit, z_fit, 'black')

# Subplot 3
ax3 = fig.add_subplot(2, 2, 3, projection='3d')
# Set axis labels and font color
ax3.set_xlabel('x', color='maroon')
ax3.set_ylabel('y', color='maroon')
ax3.set_zlabel('z', color='maroon')
# Change viewing angle
ax3.view_init(37, 45)
# Plot the data with the fit line
ax3.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)
ax3.plot3D(x_fit, y_fit, z_fit, 'black')

# Subplot 4
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
# Set axis labels and font color
ax4.set_xlabel('x', color='maroon')
ax4.set_ylabel('y', color='maroon')
ax4.set_zlabel('z', color='maroon')
# Change viewing angle
ax4.view_init(20, 26)
# Plot the data with the fit line
ax4.scatter3D(xdata, ydata, zdata, c=zdata, cmap=cm.jet)
ax4.plot3D(x_fit, y_fit, z_fit, 'black')
```
